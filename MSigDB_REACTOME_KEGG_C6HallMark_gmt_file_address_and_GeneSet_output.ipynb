### cancer type

```python
driverdb_group_cancertype = driverdb.groupby('cancer_type_abbr')
print(driverdb_group_cancertype)
```

```python
GeneList_from_cancer_type_count_ToolPredictive = []
number_from_cancer_type_count = []
gene_set_from_cancer_type = []
for i in driverdb_group_cancertype.first().index:
    c_number = []
    tmp_2 = []
    cancer_type = []
    tmp = ((driverdb_group_cancertype.get_group(i).driver_gene).str.split(', ')).tolist()
    for content in tmp:
        tmp_2.extend(content)
        cancer_type.append(content)
        c_number.append(len(content))
    tmp_2 = set(tmp_2)
    gene_set_from_cancer_type.append(cancer_type)
    number_from_cancer_type_count.append(c_number)
    GeneList_from_cancer_type_count_ToolPredictive.append(tmp_2)
print(len(GeneList_from_cancer_type_count_ToolPredictive))
print(len(number_from_cancer_type_count))
print(len(GeneList_from_cancer_type_count_ToolPredictive[0]))
print(len(gene_set_from_cancer_type[0]))
print(len(gene_set_from_cancer_type[0][0]))
print(len(gene_set_from_cancer_type[0][4]))
print(len(gene_set_from_cancer_type))
```

```python
#print(GeneList_from_cancer_type_count_ToolPredictive[0])
with open('D://downloads/node_set.txt','a') as gene_set:
    for i in range(len(GeneList_from_cancer_type_count_ToolPredictive)):
        print(len(GeneList_from_cancer_type_count_ToolPredictive[i]))
        for k in GeneList_from_cancer_type_count_ToolPredictive[i]:
            if k =='':
                continue
            gene_set.write(str(driverdb_group_cancertype.first().index[i]))
            gene_set.write('\t')
            gene_set.write(k)
            gene_set.write('\n')
```
